# Final Grade Prediction Project

# Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the Dataset

```python
dataset = pd.read_csv('student-mat.csv', sep= ';')
X = dataset[
    [
        "G1","G2","studytime","failures","absences",
        "age","sex","address","Medu","Fedu",
        "famrel","freetime","goout","health"
    ]
]
y = dataset["G3"]
```

```python
print(X)
```

# Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 0)
```

# Encoding Categorical Data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
categorical_features = ["sex", "address"]
ct = ColumnTransformer(transformers= [('encoder', OneHotEncoder(), categorical_features)], remainder= 'passthrough')
X_train_encoded = ct.fit_transform(X_train)
X_test_encoded = ct.fit_transform(X_test)

X_train_encoded = np.array(X_train_encoded)
X_test_encoded = np.array(X_test_encoded)
```

```python
print(X_train_encoded)
```

```python
print(X_test_encoded)
```

# Training the Multiple Linear Regression Model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train_encoded, y_train)
```

# Predicting the Test set results

```python
y_pred = regressor.predict(X_test_encoded)
np.set_printoptions(precision= 2)
y_test_array = y_test.to_numpy()
print(np.concatenate(
    (y_pred.reshape(len(y_pred), 1), y_test_array.reshape(len(y_test_array), 1)),
    axis=1
))
```
